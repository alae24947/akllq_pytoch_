```python
import torch

# Créer des tensors
a = torch.randn(3, 4)
b = torch.randn(3, 4)

# Opérations élémentaires
c = a + b                    # Addition
d = a * b                    # Multiplication element-wise
e = a @ b.T                  # Multiplication matricielle (@ ou torch.matmul)
f = torch.sum(a, dim=1)      # Somme sur une dimension

# Reshape et manipulation
g = a.view(12)               # Aplatir (vue, pas copie)
h = a.reshape(2, 6)          # Redimensionner (peut copier)
i = a.unsqueeze(0)           # Ajouter une dimension : (1, 3, 4)
j = i.squeeze()              # Retirer les dimensions de taille 1

# Indexation avancée
k = a[0]                     # Première ligne
l = a[:, 1]                  # Deuxième colonne
m = a[a > 0]                 # Éléments positifs

# Informations sur le tensor
print(f"Shape: {a.shape}")
print(f"Device: {a.device}")
print(f"Dtype: {a.dtype}")

# Transfert GPU (si disponible)
if torch.cuda.is_available():
    a_gpu = a.cuda()         # ou a.to('cuda')
    print(f"GPU: {a_gpu.device}")
```
